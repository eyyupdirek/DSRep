```python
import pandas as pd
import datetime,holidays
from datetime import date as d
def is_ecb_holiday_or_weekend():
    return  d.today() in holidays.ECB() or d.today().weekday() >=5 


url = 'https://www.emmi-benchmarks.eu/assets/modules/rateisblue/file_processing/publication/processed/hist_EURIBOR_'+ str(d.today().year) + '.csv'
df = pd.read_csv(url)
df.columns.values[0] = 'period'
df.set_index('period',inplace = True)
df.index = 'EURIBOR '+ df.index
df = df.stack().reset_index()

url= 'https://www.emmi-benchmarks.eu/assets/modules/rateisblue/file_processing/publication/processed/hist_EONIA_'+str(d.today().year)+'.csv'
ef = pd.read_csv(url)
ef = ef.iloc[:1,1:].rename(index = {0:'EONIA (Euro OverNight Index Average)'})
ef = ef.stack([0]).reset_index()

df = pd.concat([df, ef], ignore_index = True)
df.rename(columns = { df.columns[0]: "period",df.columns[1]: "date",df.columns[2]: "interest_rate" }, inplace = True)
df['date'] = pd.to_datetime(df['date'],format = '%d/%m/%Y') 
df.set_index("date",inplace = True) 

df.index.rename('date',inplace = True)
df.reset_index(inplace = True)
max_date =  d.today() if is_ecb_holiday_or_weekend() else max(df.date)
idx = pd.date_range(min(df.date), max_date)

idx=pd.MultiIndex.from_product([idx,df.period.unique()], names=['date', 'period'])
df=df.set_index(['date','period']).reindex(idx.to_frame(index = False) [:-1]).reset_index()
df['interest_rate']=df.groupby(['period'], sort=False)['interest_rate'].apply(lambda x: x.ffill())
df=df.groupby(['date'], sort=False).tail(6)[-30:] # 6 rate types and last 5 days 
df


import pandas as pd
import datetime,holidays
from datetime import date as d
def is_ecb_holiday_or_weekend():
    return  d.today() in holidays.ECB() or d.today().weekday() >=5 


url = 'https://www.emmi-benchmarks.eu/assets/modules/rateisblue/file_processing/publication/processed/hist_EURIBOR_'+ str(d.today().year) + '.csv'
df = pd.read_csv(url)
df.columns.values[0] = 'period'
df.set_index('period',inplace = True)
df.index = 'EURIBOR '+ df.index
df = df.stack().reset_index()

url= 'https://www.emmi-benchmarks.eu/assets/modules/rateisblue/file_processing/publication/processed/hist_EONIA_'+str(d.today().year)+'.csv'
ef = pd.read_csv(url)
ef = ef.iloc[:1,1:].rename(index = {0:'EONIA (Euro OverNight Index Average)'})
ef = ef.stack([0]).reset_index()

df = pd.concat([df, ef], ignore_index = True)
df.rename(columns = { df.columns[0]: "period",df.columns[1]: "date",df.columns[2]: "interest_rate" }, inplace = True)
df['date'] = pd.to_datetime(df['date'],format = '%d/%m/%Y') 
df.set_index("date",inplace = True) 

df.index.rename('date',inplace = True)
df.reset_index(inplace = True)

max_date =  d.today() if is_ecb_holiday_or_weekend() else max(df.date)
idx = pd.date_range(min(df.date), max_date)

idx = pd.MultiIndex.from_product([idx,df.period.unique()], names=['date', 'period'])
df = df.set_index(['date','period']).reindex(idx.to_frame(index = False) [:-1]).reset_index()
df['interest_rate'] = df.groupby(['period'], sort=False)['interest_rate'].apply(lambda x: x.ffill())
df=df.groupby(['date'], sort=False).tail(6)[-30:] # 6 rate types and last 5 days 
df
```
